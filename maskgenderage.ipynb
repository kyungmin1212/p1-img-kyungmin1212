```python
import os
import pandas as pd
from PIL import Image
import numpy as np
import matplotlib.pyplot as plt

mask_path='/opt/ml/realcode/output/mask1.csv'
age_path='/opt/ml/realcode/output/age1.csv'
gender_path='/opt/ml/realcode/output/gender1.csv'
```

```python
mask_pd=pd.read_csv(mask_path)
age_pd=pd.read_csv(age_path)
gender_pd=pd.read_csv(gender_path)
```

```python
mask_list=list(mask_pd['ans'])
age_list=list(age_pd['ans'])
gender_list=list(gender_pd['ans'])
```

```python
print(len(mask_list))
print(len(age_list))
print(len(gender_list))
```

```python
answerlist=[]
for i in range(len(mask_list)):
    result=mask_list[i]*6+gender_list[i]*3+age_list[i]
    answerlist.append(result)
```

```python
info_path='/opt/ml/input/data/eval/info.csv'
info = pd.read_csv(info_path)

info['ans']=answerlist
info.to_csv('/opt/ml/realcode/output/maskgenderage1.csv',index=False)
```
